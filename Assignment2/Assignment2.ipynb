```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Admission_Predict.csv")
data.head()
```

```python
data.isnull().sum().sum()
```

```python
print(f"statistics of data \n {data.describe()}")
```

```python
data2 = data.drop(columns=['Serial No.'])
data2.hist(bins=10, figsize=(15, 10))
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(15, 14))
for i, column in enumerate(data2.columns, 1):
    plt.subplot(3, 3, i)
    sns.kdeplot(data2[column])
    plt.title(f'Distribution of {column}')
```
